## Data preprcoessing

```python
# Split the data into training and testing sets
X = df.drop('Class', axis=1)
y = df['Class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=info_cluster, random_state=42)
```

```python
# Check for outliers using Local Outlier Factor (LOF) on X_train
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
outliers = lof.fit_predict(X_train)
outlier_indices = np.where(outliers==-1)
X_train = X_train.drop(X_train.index[outlier_indices], axis=0)
y_train = y_train.drop(y_train.index[outlier_indices])
print(f"Number of outlier records removed: {len(outlier_indices[0])}")
```

```python
class_counts = y_train.value_counts()
plt.bar(class_counts.index, class_counts.values)
plt.title('Class distribution in y_train')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

## Feature reduction

```python
from boruta import BorutaPy
```

```python
X_train.shape, y_train.shape
```

```python
rf = RandomForestClassifier(n_jobs=-1, random_state=42)
boruta_selection = BorutaPy(rf, n_estimators='auto', verbose=3, random_state=42)
boruta_selection.fit(X_train.values, y_train.values)
```

```python
# Get the selected features
selected_features = X_train.columns[boruta_selection.support_].tolist()
print(f"Selected features: {selected_features}")
```

## Class balancing

```python
oversampler = RandomOverSampler(random_state=42)
X_train_boruta = X_train[selected_features]
X_train_resampled, y_train_resampled = oversampler.fit_resample(X_train_boruta, y_train)
df_resampled = pd.concat([X_train_resampled, y_train_resampled], axis=1)
```

```python
X_train_resampled.shape, y_train_resampled.shape
```

```python
class_counts = y_train_resampled.value_counts()
plt.bar(class_counts.index, class_counts.values)
plt.title('Class distribution in y_train')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

## Feature scaling

```python
sc = StandardScaler()
X_train_scaler = sc.fit_transform(X_train_resampled)
```

## Modelling

### Parameter hypertunining for random forest

```python
# Train the random forest classifier with the selected features
rf = RandomForestClassifier(random_state=42)
```

```python
# Define the parameter grid for the grid search
param_grid = {
    'n_estimators': [100, 200, 500],
    'max_features': ['sqrt', 'log2'],
    'max_depth': [10, 20, None]
}
```

```python
grid_search = GridSearchCV(rf, param_grid=param_grid, cv=5, n_jobs=-1, scoring='roc_auc')
grid_search.fit(X_train_scaler, y_train_resampled)
best_params = grid_search.best_params_
print(f"Best parameters: {best_params}")
```

```python
best_rf_model = grid_search.best_estimator_

# Evaluate the model on the training set
y_train_pred = best_rf_model.predict(X_train_scaler)
y_train_prob = best_rf_model.predict_proba(X_train_scaler)[:, 1]

print("Training set metrics:")
print(f"ROC AUC score: {roc_auc_score(y_train_resampled, y_train_prob)}")
print(f"Accuracy: {accuracy_score(y_train_resampled, y_train_pred)}")
print(f"Precision: {precision_score(y_train_resampled, y_train_pred)}")
print(f"Recall: {recall_score(y_train_resampled, y_train_pred)}")
print(f"F1-score: {f1_score(y_train_resampled, y_train_pred)}")
```

```python
import xgboost as xgb
```

```python
y_resampled_for_xgb = y_train_resampled.copy(deep=True)
y_resampled_for_xgb[y_resampled_for_xgb == -1] = 0
```

```python
xgb_classifier = xgb.XGBClassifier()
```

```python
param_grid = {
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 1],
    'n_estimators': [50, 100, 200]
}
```

```python
# Perform grid search using 5-fold cross validation
grid_search = GridSearchCV(xgb_classifier, param_grid, cv=5, scoring='roc_auc', n_jobs=-1)

# Fit the grid search to the resampled training data
grid_search.fit(X_train_scaler, y_resampled_for_xgb)

# Print the best hyperparameters and the corresponding ROC AUC score
print("Best Parameters: ", grid_search.best_params_)
print("Best Score: ", grid_search.best_score_)
```

```python
xgb_classifier = grid_search.best_estimator_
xgb_classifier.fit(X_train_scaler, y_resampled_for_xgb)
y_train_predictions_for_xgb = xgb_classifier.predict(X_train_scaler)

print("Training set metrics:")
print(f"ROC AUC score: {roc_auc_score(y_resampled_for_xgb, y_train_predictions_for_xgb)}")
print(f"Accuracy: {accuracy_score(y_resampled_for_xgb, y_train_predictions_for_xgb)}")
print(f"Precision: {precision_score(y_resampled_for_xgb, y_train_predictions_for_xgb)}")
print(f"Recall: {recall_score(y_resampled_for_xgb, y_train_predictions_for_xgb)}")
print(f"F1-score: {f1_score(y_resampled_for_xgb, y_train_predictions_for_xgb)}")
```

##  Prediction on test set

```python
test_df = pd.concat([X_test, y_test], axis=1)
test_df = test_df.dropna()
```

```python
y_test = test_df.pop('Class')
X_test = test_df
X_test.shape, y_test.shape
```

```python
X_test = X_test[selected_features]
X_test_scaler = sc.transform(X_test)

# Evaluate the model on the training set
y_test_pred = best_rf_model.predict(X_test_scaler)
y_test_prob = best_rf_model.predict_proba(X_test_scaler)[:, 1]

print("Testing set metrics for Random Forest Classifier:")
print(f"ROC AUC score: {roc_auc_score(y_test, y_test_prob)}")
print(f"Accuracy: {accuracy_score(y_test, y_test_pred)}")
print(f"Precision: {precision_score(y_test, y_test_pred)}")
print(f"Recall: {recall_score(y_test, y_test_pred)}")
print(f"F1-score: {f1_score(y_test, y_test_pred)}")
```

```python
y_test_for_xgb = y_test.copy(deep=True)
y_test_for_xgb[y_test_for_xgb == -1] = 0

y_test_predictions = xgb_classifier.predict(X_test_scaler)
print("Test set metrics for XGB:")
print(f"ROC AUC score: {roc_auc_score(y_test_for_xgb, y_test_predictions)}")
print(f"Accuracy: {accuracy_score(y_test_for_xgb, y_test_predictions)}")
print(f"Precision: {precision_score(y_test_for_xgb, y_test_predictions)}")
print(f"Recall: {recall_score(y_test_for_xgb, y_test_predictions)}")
print(f"F1-score: {f1_score(y_test_for_xgb, y_test_predictions)}")
```

```python
auc_scores = [roc_auc_score(y_test, y_test_prob), roc_auc_score(y_test_for_xgb, y_test_predictions)]

# Define the model names
model_names = ['Random Forest', 'XGBoost']

# Create a bar chart of the AUC scores
plt.bar(model_names, auc_scores)
plt.title('Performance Comparison of Models')
plt.xlabel('Model')
plt.ylabel('AUC Score')
plt.show()
```

The result of this code shows a bar chart with two bars, where the first bar represents the AUC score of the Random Forest model, and the second bar represents the AUC score of the XGBoost model.
